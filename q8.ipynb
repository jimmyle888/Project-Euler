```python
def greatestProduct(series, numDigits):
    maxProduct = 1
    for i in range(0, len(series)-(numDigits-1)):
        tempSeries = series[i:i+numDigits]
        #print(tempSeries)
        product = 1
        for j in tempSeries:
            product = product * int(j)
        maxProduct = max(product, maxProduct)
    print(maxProduct)

        
    
```
