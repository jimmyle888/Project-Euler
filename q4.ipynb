```python
def isPalindrome(num):
    #flag = True
    numStr = str(num)
    l = 0
    r = len(numStr) - 1
    while (l < r):
        # print("l=" + str(l) + " " + str(numStr[l]))
        # print("r=" + str(r) + " " + str(numStr[r]))
        if numStr[l] != numStr[r]:
            #print("hi")
            return False 
        l = l + 1
        r = r - 1 
    return True
```

```python
maxPalindrome = 0

for i in range(100,999):
    for j in range(100, 999):
        product = i * j
        if isPalindrome(product):
            maxPalindrome = max(product, maxPalindrome)

print(maxPalindrome)
        
        
```
