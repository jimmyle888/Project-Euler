```python
def pythagTriplet(sln):
    for c in range (1, sln):
        for b in range (1, c):
            for a in range (1, b):
                if (a**2 + b**2 == c**2 and a + b + c == 1000):
                    print("sol= a=%d b=%d c=%d" %(a,b,c))
                    A = a
                    B = b
                    C = c
                    return (A, B, C)
                
[A, B, C] = pythagTriplet(1000)

print(A * B * C)
```
