```python
def isPrime(n):
    flag = True
    n = int(n)
    for i in range(2,n): #Loops all numbers up to n-1
        if n % i == 0:
            flag = False
            break
    return flag
        
print(isPrime(4))
```

```python
#Return the largest prime number factor of a given n
def getPrimeNumbers(n):
    primes = []
    n = int(n)
    for i in range(2,n):
        if (isPrime(i)):
            primes.append(i)
    print(primes)
    return primes
```

```python
getPrimeNumbers(100)
```

```python
def getLargestPrimeFactor(num):
    #print("hi")
    primes = getPrimeNumbers(num)
    if isPrime(num):
        return num
    else:
        ##Find some code to divide by smallest prime numbers
        for i in primes:
            if (num % i == 0):
                factor1 = num / i
                factor2 = num / factor1
                #print("factor1=" + str(factor1))
                #print("factor2=" + str(factor2))
                break
        return max(getLargestPrimeFactor(factor1), getLargestPrimeFactor(factor2))

```
